```python
import sys
import os

sys.path.append(os.path.abspath("../..")) #make the folder "automl" part of this
```

```python
experiment_path = 'data\\experiments\\HyperparameterOptimizationPipeline_13'
```

```python
configuration_path = 'configuration_1'
```

```python
results_path = 'results.csv'
#results_path = 'agent_1\\results.csv'
#results_path = 'agent_2\\results.csv'
```

```python
from automl.loggers.result_logger import ResultLogger

results_to_look = ResultLogger(input={
                                        "logger_directory" : f"{experiment_path}\\{configuration_path}",
                                        "filename" : results_path
                                      })
results_to_look.proccess_input()

print(f"Opened results with columns: {results_to_look.columns}")
```

```python
results_to_look.plot_graph(x_axis='episode', y_axis=['total_reward'])
results_to_look.plot_confidence_interval(x_axis='episode', y_column='total_reward')
```
